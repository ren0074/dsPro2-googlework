### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
origin_string = "datascience"

reversed_string = origin_string[::-1]

print(f"元の文字列: {origin_string}")
print(f"逆順の文字列: {reversed_string}")
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
origin_string = "ヘエルスシトンニキア"

extracted_chars = origin_string[::2]

print(f"元の文字列: {origin_string}")
print(f"抽出された文字列: {extracted_chars}")
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
string1 = "ヘルシンキ"
string2 = "エストニア"

merged_chars = []

for char1, char2 in zip(string1, string2):
    merged_chars.append(char1)
    merged_chars.append(char2)

result_string = "".join(merged_chars)

print(f"文字列1: {string1}")
print(f"文字列2: {string2}")
print(f"交互に連結した文字列: {result_string}")
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"
words = sentence.split()
word_lengths = []

for word in words:
    word_lengths.append(len(word))

print(f"元の文: {sentence}")
print(f"単語リスト: {words}") # オプション：単語リストも確認したい場合
print(f"各単語の文字数リスト: {word_lengths}")
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."

words = sentence.split()

target_positions = [1, 2, 5, 6, 8, 9, 16]

result_dict = {}

for i, word in enumerate(words):

    current_position = i + 1

    if current_position in target_positions:
       
        first_two_chars = word[:2]
        
        result_dict[first_two_chars] = current_position

print(f"元の文: {sentence}")
print(f"分解された単語リスト (一部): {words[:10]}...") #多いため省略
print(f"抽出された単語の先頭2文字と位置の辞書: {result_dict}")
```
